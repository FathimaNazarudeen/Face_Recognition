```python
import cv2
import numpy
import os 
import face_recognition
```

```python
path='images'
```

```python
mylist=os.listdir(path)
mylist
```

```python
imgs=[]
classnames=[]
for img in mylist:
    imgpath=os.path.join(path,img)
    current_img=cv2.imread(imgpath)
    print(current_img)
    imgs.append(current_img)
    classnames.append(img.split('.')[0])
print(len(imgs))
print(classnames)
```

```python
def faceencoding(images):
    encodelist=[]
    for img in images:
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        face_in_frame=face_recognition.face_locations(img)
        #print(face_in_frame)
        encode=face_recognition.face_encodings(img,face_in_frame)[0]
        #print(encode)
        encodelist.append(encode)
    return encodelist
encode_list_known_faces=faceencoding(imgs)
print(encode_list_known_faces)
```

```python
import numpy as np
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    face_in_frame=face_recognition.face_locations(img1)
    #print(face_in_frame)
    encoded_face=face_recognition.face_encodings(img1,face_in_frame)
    #print(encoded_face)
    for enc_face,faceloc in zip(encoded_face,face_in_frame):
        matches=face_recognition.compare_faces(encode_list_known_faces,enc_face)
        #print(matches)
        face_dist=face_recognition.face_distance(encode_list_known_faces,enc_face)
        #print(face_dist)
        matchind=np.argmin(face_dist)
        #print(matchind)
        if matches[matchind]==True:
            name=classnames[matchind]
            print(name)
            y1,x2,y2,x1=faceloc
            cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,255),4)
            cv2.putText(img,name,(x1,y1),cv2.FONT_HERSHEY_COMPLEX,fontScale=4,color=(255,0,234),thickness=10)
    cv2.imshow('IMAGE',img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
    
```
